```python
import matplotlib.pyplot as plt
import numpy as np

product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales= np.array([200,300,500])



plt.bar(product_category, sales)
plt.title("this si title\n", fontdict={'fontsize': 20, 'fontweight':300, 'color':'purple'} )

plt.xlabel("\nThings\n", fontdict={'fontsize':20, 'fontweight': 100, 'color': "blue"} )
plt.ylabel("Quantity", fontdict={'fontsize':20, 'fontweight': 100, 'color': "blue"} )

ticks=(np.arange(1,600,50))
label=["{}M".format(i/100) for i in ticks]

plt.yticks(ticks, label)
plt.xticks(rotation=90)

plt.show()
```
